```python
def extract_numeric(value):
    if isinstance(value, str):
        # Use regular expression to find the numerical part
        matches = re.findall(r'\d*\.?\d+', value)
        # Check if matches exist
        if matches:
            num_part = matches[0]
            # Convert 'k' to thousand
            if 'k' in value:
                return float(num_part) * 1000
            else:
                return float(num_part)
        else:
            return 0
    else:
        return value

# Apply the function to the column
df['favorite'] = df['favorite'].apply(extract_numeric)
df.favorite.unique()
```

# **5. RFM Analysis**

```python
import datetime as dt

# Ensure w_date is in datetime format
data_cleaned['w_date'] = pd.to_datetime(data_cleaned['w_date'])

# Define the reference date (e.g., 2024-12-31)
reference_date = dt.datetime(2024, 12, 31)

# Calculate the most recent transaction date for each customer
recency = data_cleaned.groupby('id')['w_date'].max().reset_index()

# Calculate Recency as the number of days since the last transaction
recency['Recency'] = (reference_date - recency['w_date']).dt.days

# Merge Recency back into the main DataFrame (if needed)
data_cleaned = data_cleaned.merge(recency[['id', 'Recency']], on='id', how='left')

# Display results
print(data_cleaned[['id', 'Recency']].head())
```

## **5.1. RFM Segment**

```python
# Create Recency Segments based on specified bin ranges
recency_bins = pd.cut(data_cleaned['Recency'], bins=[597.999, 599.0, 608.0, 617.0], 
                      labels=["Very Recent", "Recent", "Inactive"], include_lowest=True)
data_cleaned['Recency_Segment']=recency_bins

# Create Frequency Segments based on specified bin ranges
frequency_bins = pd.cut(data_cleaned['Frequency'], bins=[0, 203.0, 1348.667, 482000.0], 
                        labels=["Low Frequency", "Medium Frequency", "High Frequency"], include_lowest=True)
data_cleaned['Frequency_Segment']=frequency_bins

# Create Monetary Segments based on specified bin ranges
monetary_bins = pd.cut(data_cleaned['Monetary'], bins=[0, 2.47, 39.9, 1000000], 
                       labels=["Low Spend", "Medium Spend", "High Spend"], include_lowest=True)
data_cleaned['Monetary_Segment']=monetary_bins
```
